```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
import re

# Data Acquisition and Preprocessing
data = pd.read_csv('zillow data copy.csv')

# Extract date columns and sort them chronologically
date_columns = [col for col in data.columns if re.match(r'\d{4}-\d{2}-\d{2}', col)]
date_columns.sort()

# Clean data - handle missing values
data_clean = data.copy()
data_clean[date_columns] = data_clean[date_columns].ffill(axis=1)  # Forward fill
data_clean[date_columns] = data_clean[date_columns].bfill(axis=1)  # Backward fill
# Drop any remaining rows with NaNs
data_clean = data_clean.dropna(subset=date_columns)

# Create sequences of data for RNN
sequence_length = 12  # Use 12 months of data to predict the next month
features = data_clean[date_columns].values

# Transform features to be between 0 and 1 (scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# Create time series sequences with proper dimensions
def create_sequences(data, seq_length):
    X, y = [], []
    # Each row is a region, columns are time steps
    for i in range(data.shape[0]):  # For each region
        region_data = data[i, :]
        # Create sequences for this region
        for j in range(len(region_data) - seq_length):
            X.append(region_data[j:j+seq_length])
            y.append(region_data[j+seq_length])
    return np.array(X), np.array(y)

# Create single dataset from all regions
X, y = create_sequences(features_scaled, sequence_length)

# Check shapes to ensure they match
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")

# Reshape for RNN 
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split into training and testing sets (80-20 split)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
# Build the RNN model
model = keras.Sequential([
    layers.Input(shape=(sequence_length, 1)),
    layers.LSTM(64, return_sequences=True),
    layers.Dropout(0.1),
    layers.LSTM(32),
    layers.Dropout(0.1),
    layers.Dense(16, activation='relu'),
    layers.Dense(1)
])

# Compile with appropriate loss function and optimizer for regression
optimizer = keras.optimizers.Adam(learning_rate=0.001, clipnorm=1.0)
model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

# Add callbacks for better training
callbacks = [
    keras.callbacks.EarlyStopping(patience=15, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5, min_lr=1e-5),
    keras.callbacks.ModelCheckpoint('best_rnn_model.keras', save_best_only=True)
]

# Train the model
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=callbacks,
    verbose=2
)
```

```python
# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(122)
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Model MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.tight_layout()
plt.show()

# Evaluate on test set
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f"Test MAE: {test_mae:.4f}")

# Make predictions and plot actual vs predicted
y_pred = model.predict(X_test)
```

```python

# Inverse transform to get original values
y_test_scaled = y_test.reshape(-1, 1)
y_pred_scaled = y_pred

# Get the min and max values used in the original scaling
data_min = scaler.data_min_
data_max = scaler.data_max_
data_range = data_max - data_min

# Apply inverse scaling to the specific column we're predicting
y_test_original = y_test_scaled * data_range[0] + data_min[0]
y_pred_original = y_pred_scaled * data_range[0] + data_min[0]

# Plot actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test_original, y_pred_original, alpha=0.5)
min_val = min(y_test_original.min(), y_pred_original.min())
max_val = max(y_test_original.max(), y_pred_original.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2)
plt.xlabel('Actual House Values ($)')
plt.ylabel('Predicted House Values ($)')
plt.title('Actual vs Predicted House Values')
plt.tight_layout()
plt.show()

# Calculate additional metrics
from sklearn.metrics import mean_squared_error, r2_score
rmse = np.sqrt(mean_squared_error(y_test_original, y_pred_original))
r2 = r2_score(y_test_original, y_pred_original)
print(f"Root Mean Squared Error: ${rmse:.2f}")
print(f"R-squared: {r2:.4f}")
```

```python

# Error analysis
errors = y_pred_original - y_test_original
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.hist(errors, bins=30)
plt.title('Error Distribution')
plt.xlabel('Prediction Error ($)')
plt.ylabel('Frequency')

plt.subplot(122)
plt.scatter(y_test_original, errors, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Error vs Actual Value')
plt.xlabel('Actual House Values ($)')
plt.ylabel('Prediction Error ($)')
plt.tight_layout()
plt.show()

print("\nModel Limitations and Real-World Implications:")
print("1. The model predicts based solely on historical price patterns without considering")
print("   external factors like economic indicators, policy changes, or local development.")
print("2. The model assumes temporal patterns are consistent across different regions,")
print("   which may not hold true in practice.")
print("3. The prediction horizon is limited to one month ahead; longer forecasts would")
print("   likely accumulate errors.")
print("4. Housing markets often experence strucural changes that historical data cannot capture.")
```
